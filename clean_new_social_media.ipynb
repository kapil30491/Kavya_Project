```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Social_Media.csv")
df
```

```python
df.info()
```

```python
# Handle null Values Using fillna
df["likes"] = df["likes"].fillna(df["likes"].median())
```

```python
df.info()
```

```python
# change datatype
# change data type 
df['likes'] = df['likes'].astype(int)
df['date'] = pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
# save data
df.to_csv("clean_social_media.csv", index=False)
```
