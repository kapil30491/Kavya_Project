```python
df=pd.read_csv('clean_campaign_data.csv')
df
```

```python
df['spend'] = df['spend'].astype(int)
```

```python
df.info()
```

```python
# Partywise total spending
party_spending = df.groupby('party')['spend'].sum().reset_index()
print(party_spending)
```

```python
party_spending = party_spending.sort_values(by="spend",ascending=False)
print(party_spending)
```

```python
# BJP had the highest campaign spending, investing approximately $6.61 billion
# Other Major Contributors:
# Regional Party spent $6.21 billion closely followed by INC with $6.06 billion.
# AAP spent the least among the listed parties, totaling $5.86 billion.
# Spending Distribution:
# The campaign spending is fairly balanced among all four parties, with each party contributing roughly one-fourth of the total spending.
# This indicates a competitive campaign environment, with substantial investments by all major participants.

# Insights:
# The slight differences in spending may reflect strategic allocation to key states or channels.
# Parties with higher spending might have invested in more outreach programs, advertisements, or events to influence voter engagement.
# Such data can help in assessing campaign intensity and resource utilization across parties.

```

```python
# Statewise total spending

state_spending = df.groupby('state')['spend'].sum().reset_index()
print(state_spending)
```

```python
# Statewise AVERAGE spending

state_spending = df.groupby('state')['spend'].mean().reset_index()
print(state_spending)
```

```python
state_spending_sorted = state_spending.sort_values(by="spend", ascending=False)
print(state_spending_sorted)
```

```python
# Highest Spending:
# Uttar Pradesh (UP) received the highest campaign spending, totaling approximately $6.60 billion. This indicates that parties heavily invested
# resources in this state, likely due to its large voter base and political significance.
# Moderate Spending:
# Tamil Nadu had the second-highest spending at $6.32 billion, followed by Bihar with $5.95 billion. These states also represent strategically 
# important regions for the campaign.
# Lowest Spending:
# Maharashtra recorded the lowest among the four states at $5.88 billion, though the amount is still substantial.
# Balanced Investment:
# The difference in spending across the states is not extreme, suggesting that parties aimed for a relatively balanced distribution of campaign
# resources across key regions.
```

# Insights for Strategic Decision Making:

```python
Resource Allocation: The higher spending in UP and Tamil Nadu reflects targeted resource allocation in states with higher voter turnout
or competitive constituencies.
Campaign Prioritization: Parties may have prioritized states based on electoral significance, past performance, or voter density.
Efficiency Analysis: Comparing spending with votes gained or seats won in each state could provide deeper insight into campaign ROI and effectiveness
of resource utilization.
```
