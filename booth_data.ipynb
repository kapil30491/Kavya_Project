```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Booth_Level_Data.csv')
df
```

```python
# check this
df.isnull().sum()
```

```python
df["status"] = np.where((df["registered_voters"].isna()) | (df["votes_cast"] > df["registered_voters"]), "INVALID", "VALID")
```

```python
df
```

```python
df["Actual votes cast"] = np.where(df["registered_voters"].isna(), 0, df[["votes_cast","registered_voters"]].min(axis=1))
```

```python
df
```

```python
df["registered_voters"] = df["registered_voters"].fillna(0).astype(int)
```

```python
df.info()
```

```python
df["registered_voters"] = df["registered_voters"].astype("int64")
```

```python
df.info()
```

# Now data is cleaned format and ready for Calculations

```python
df.to_csv("cleaned_booth_data.csv", index=False)
```
